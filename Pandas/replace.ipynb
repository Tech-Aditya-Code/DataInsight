```python
var.replace(to_replace=["Computer"],value="Python")
```

```python
var.replace([1,2,3,4,5,6,7,8,9],22)
```

```python
var.replace({"Name":'[A-Z]'},22,regex=True)
```

```python
var.replace(1,method="bfill")
```
